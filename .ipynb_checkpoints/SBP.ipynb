```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
superbowl_csv = "../Superbowl-Presidents/Data/superbowl.csv"
superbowl_df = pd.read_csv(superbowl_csv)
superbowl_df.head()
```

```python
president_tab = "../Superbowl-Presidents/Data/president.tab"
president_df = pd.read_table(president_tab)
president_df.head()

```

```python
superbowl_df= superbowl_df[["Date", "Winner", "Loser", "SB"]]
superbowl_df['Year'] = pd.DatetimeIndex(superbowl_df['Date']).year
superbowl_df.columns = map(str.lower, superbowl_df.columns)
superbowl_df.head()
```

```python
president_1967 = president_df.loc[(president_df['year']>1966)]
president_1967['total_votes'] = president_1967.groupby(['year', 'party_detailed', 'candidate'])['candidatevotes'].transform('sum')
president_winner = president_1967.loc[president_1967.groupby(['year'])['total_votes'].idxmax()]
president_winner = president_winner[['year','candidate','party_detailed']]
president_winner.head()
```

```python
from config import postgres_credentials as creds
rds_connection_string = f"{creds['user']}:{creds['pass']}@localhost:5432/SBP"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
president_winner.to_sql(name="presidents", con=engine, if_exists='append', index=False)
```

```python
superbowl_df.to_sql(name="superbowl", con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('SELECT s.date,s.winner, s.loser, s.sb, p.candidate, p.party_detailed FROM superbowl s JOIN presidents p ON s.year = p.year ', con=engine).head(20)
```
