```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
superbowl_csv = "../Project 2/Data/superbowl.csv"
superbowl_df = pd.read_csv(superbowl_csv)
superbowl_df.head()
```

```python
president_csv = "../Project 2/Data/presidents.csv"
president_df = pd.read_csv(president_csv)
president_df.head()

```

```python
superbowl_df= superbowl_df[["Date", "Winner", "Loser", "SB"]]
superbowl_df['Year'] = pd.DatetimeIndex(superbowl_df['Date']).year
superbowl_df.columns = map(str.lower, superbowl_df.columns)
superbowl_df.head()
```

```python
president_df= president_df[["President","Took_office", "Party"]]
president_df['Year'] = pd.DatetimeIndex(president_df['Took_office']).year-1
president_df.columns = map(str.lower, president_df.columns)
president_df.head()
```

```python
from config import postgres_credentials as creds
rds_connection_string = f"{creds['user']}:{creds['pass']}@localhost:5432/SBP"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
president_df.to_sql(name="presidents", con=engine, if_exists='append', index=False)
```

```python
superbowl_df.to_sql(name="superbowl", con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('SELECT s.date,s.winner, s.loser, s.sb, p.president, p.party FROM superbowl s JOIN presidents p ON s.year = p.year ', con=engine).head(20)
```
